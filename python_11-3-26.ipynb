```python
# List of scores
scores = [78, 40, 92, 34, 67, 88, 53, 41, 76, 29, 64, 81, 39, 55, 72, 90, 47, 68, 33, 84]

# 2. Slicing
first_10 = scores[:10]
last_5 = scores[-5:]
every_second = scores[::2]

print("First 10 scores:", first_10)
print("Last 5 scores:", last_5)
print("Every second score:", every_second)
```

```python
# 3. Iteration (Pass / Fail)
print("\nStudent Results:")
for score in scores:
    if score >= 40:
        print(score, "- Passed")
    else:
        print(score, "- Failed")
```

```python
# 4. List Comprehension
passing_scores = [s for s in scores if s >= 40]
bonus_scores = [s + 5 for s in scores]
scores_above_75 = [s for s in scores if s > 75]

print("\nPassing scores:", passing_scores)
print("Scores with +5 bonus:", bonus_scores)
print("Scores greater than 75:", scores_above_75)

# 5. Count pass and fail
pass_count = len([s for s in scores if s >= 40])
fail_count = len([s for s in scores if s < 40])

print("\nNumber of students passed:", pass_count)
print("Number of students failed:", fail_count)
```
